```python
import pandas as pd

# Load the data
file_path = '/content/drive/MyDrive/! paper Right Wing/data/truthsocial_common_urls_6825.csv'
data = pd.read_csv(file_path)

# Display basic information and first few rows of the data
data_info = data.info()
data_head = data.head()

(data_info, data_head)
```

```python
from bs4 import BeautifulSoup

# Function to extract and strip text from HTML content
def extract_strip_text(html_content):
    try:
        soup = BeautifulSoup(html_content, 'html.parser')
        return soup.get_text(strip=True).strip()
    except TypeError:
        return None

# Apply the revised function to the 'content' column and overwrite the 'text_extract' column
data['text_extract'] = data['content'].apply(extract_strip_text)

# Display the first few rows to verify the revised extraction
data[['content', 'text_extract']].head()
```

```python
import re

def process_text_keep_hyphen(text):
    try:
        # Convert text to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'http\S+|www\.\S+', '', text)

        # Remove special characters and punctuation using regex, but keep hyphenated words
        text = re.sub(r'(?<!\w)-(?!\w)|[^\w\s-]', '', text)

        return text
    except TypeError:
        return None

# Apply the revised function to the 'text_extract' column and overwrite 'text_extract_processed'
data['text_extract_processed'] = data['text_extract'].apply(process_text_keep_hyphen)

# Display the first few rows to verify the processed text
data[['text_extract', 'text_extract_processed']].head()
```

```python
# Remove the 'text' column
data_without_text = data.drop(columns=['text'])

# Save the modified data to a pickle file
pickle_file_path_without_text = '/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/truth_processed.pkl'
data_without_text.to_pickle(pickle_file_path_without_text)
```

```python
# Save the modified data to a CSV file
csv_file_path_without_text = '/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/truth_processed.csv'
data_without_text.to_csv(csv_file_path_without_text, index=False)
```

```python
import pandas as pd
import re

# Load the data with 'ISO-8859-1' encoding
file_path = '/content/drive/MyDrive/! paper Right Wing/data/tweets_common_urls_11497.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Define a function to process the text data
def process_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text, flags=re.MULTILINE)
    # Remove special characters (keeping whitespaces)
    text = re.sub(r'[^\w\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    return text

# Process the 'text' column and replace the old 'processed_text' column
data['processed_text'] = data['text'].apply(process_text)

# Define file paths for saving
csv_file_path = '/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/twitter_processed.csv'
pickle_file_path = '/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/twitter_processed.pkl'

# Save data to CSV and Pickle
data.to_csv(csv_file_path, index=False)
data.to_pickle(pickle_file_path)
```

```python
import numpy as np
import pandas as pd
import uuid
from tqdm import tqdm
import pickle
from huggingface_hub import snapshot_download
from sentence_transformers import SentenceTransformer
import torch

get_ipython().run_line_magic('pwd', '')

ts = open('/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/truth_processed.pkl','rb')
df_ts = pickle.load(ts)

tw = open('/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/twitter_processed.pkl','rb')
df_tw = pickle.load(tw)

print(df_ts.shape, df_tw.shape)
```

```python
username_dict = {'WashTimes': 'washtimes',
                 'FDRLST': 'FDRLST',
                 'RealClearNews': 'RealClearNews',
                 'NEWSMAX': 'NewsMax',
                 'BreitbartNews': 'breitbartnews',
                 'FoxNews': 'FoxNews',
                 'OANN': 'OAN',
                 'dcexaminer': 'WashingtonExaminer',
                 'theMRC': 'theMRC'}

# give twitter username the truthsocial source
df_tw['source'] = [username_dict[item] for item in df_tw.username]
print(df_tw.source.value_counts().to_dict(), df_tw.username.value_counts().to_dict())
```

```python
# create uuid for all urls
urls_all = set(df_tw.url_expanded).union(set(df_ts.url_extract))
print(len(urls_all))

uuids = [uuid.uuid4() for _ in urls_all]

# If you want to create a dictionary mapping each URL to its UUID:
url_to_uuid = dict(zip(urls_all, uuids))

# attach uuid for all urls
df_tw['url_uuid'] = [url_to_uuid.get(item) for item in df_tw.url_expanded]
df_ts['url_uuid'] = [url_to_uuid.get(item) for item in df_ts.url_extract]
# for merging purpose, add the prefix to the different platforms
df_tw = df_tw.add_prefix('tw_')
df_ts = df_ts.add_prefix('ts_')
# select cols to merge
# drop duplicated message (including urls) in each platform
df_tw1 = df_tw[['tw_source', 'tw_url_uuid', 'tw_text', 'tw_processed_text', 'tw_url_expanded']].drop_duplicates()
df_tw1['tw_source_content_uuid'] = ['tw-' + str(uuid.uuid4()) for _ in range(len(df_tw1))]
print(df_tw1.shape, df_tw.shape)

df_ts1 = df_ts[['ts_source', 'ts_url_uuid', 'ts_content', 'ts_text_extract_processed', 'ts_url_extract']].drop_duplicates()
df_ts1['ts_source_content_uuid'] = ['ts-' + str(uuid.uuid4()) for _ in range(len(df_ts1))]
print(df_ts1.shape, df_ts.shape)


# merge source and url_uuid on two platforms
# same source, same ulr, one platform may post different multiple messages
dfm = df_tw1.merge(df_ts1,
            left_on=['tw_source', 'tw_url_uuid'],
            right_on=['ts_source', 'ts_url_uuid'],
            how='inner')
print(dfm.shape)
dfm.head(1)

# give cuuid to each comparison
dfm['comparison_uuid'] = dfm.tw_source_content_uuid + "-" + dfm.ts_source_content_uuid
dfm.comparison_uuid.nunique()

```

```python
# Define file paths for saving
csv_file_path = '/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/merge_comparison.csv'
pickle_file_path = '/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/merge_comparison.pkl'

# Save data to CSV and Pickle
dfm.to_csv(csv_file_path, index=False)
dfm.to_pickle(pickle_file_path)
```

```python
# get bert similarity
# compare twitter raw text, with truth social text
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
comparisons = list(zip(dfm.tw_processed_text, dfm.ts_text_extract_processed))
comparisons_label = list(zip(dfm.tw_source_content_uuid, dfm.ts_source_content_uuid))
comparisons_result = []
for idx, sentences in enumerate(tqdm(comparisons)):
    embeddings = model.encode(sentences)
    result = np.dot(embeddings, embeddings.T)
    i, j = np.triu_indices(result.shape[0], k=1)
    upper_triangle_values = result[i, j]
    label = comparisons_label[idx]
    comparisons_result.append((label[0], label[1], upper_triangle_values[0]))
```

```python
# merge results back to dfm
df_result = pd.DataFrame(comparisons_result,
                         columns = ['tw_source_content_uuid', 'ts_source_content_uuid', 'bert_sim'])
df_result['comparison_uuid'] = df_result.tw_source_content_uuid + "-" + df_result.ts_source_content_uuid
dfm1 = dfm.merge(df_result, left_on=['tw_source_content_uuid', 'ts_source_content_uuid', 'comparison_uuid'],
                     right_on=['tw_source_content_uuid', 'ts_source_content_uuid', 'comparison_uuid'],
                     how='left')
with open("/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/twitter_truth_sim_BERT.pkl", "wb") as f:
    pickle.dump(dfm1, f)
dfm1.to_csv("/content/drive/MyDrive/! paper Right Wing/data/01 urls on both twitter and truth/twitter_truth_sim_BERT.csv", index=False)
```
