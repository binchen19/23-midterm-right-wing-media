```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data
parties = ['Democratic', 'Republican', 'Others']
strategies = ['Fundraising', 'Voting', 'Event/Rally', 'Online Engagement', 'Visit Other Platforms']
values = np.array([
    [4480, 18595, 8178, 1803, 396],
    [1323, 11295, 6598, 3888, 1089],
    [1047, 4607, 2303, 2512, 63]
])
N_values = np.array([29362, 22234, 9436])

# Calculate percentages
percentages = (values / N_values[:, None]) * 100

# Plotting
plt.figure(figsize=(10, 6))
sns.heatmap(percentages, annot=True, cmap='Blues', xticklabels=strategies, yticklabels=parties, fmt=".2f", cbar_kws={'label': 'Percentage (%)'})
plt.title("Mobilization Strategies by Party")
plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=20, ha="right")
plt.yticks(rotation=0)
plt.show()



```
